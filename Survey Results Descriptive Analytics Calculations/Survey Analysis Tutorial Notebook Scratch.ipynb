```python
## Importing Python Libraries

import pandas as pd
import numpy as ny
```

```python
df = pd.read_excel('MUD.xlsx')
df.tail()
```

```python
## Listing Column Names

list(df)
```

```python
## Converting numeric values to string variables

df['Gender'] = df['Gender'].apply(lambda x: 'Female' if x == 1 else 'Male')

print (df.Gender)
```

```python
## Frequency of values as percentage ##
## Distribution of interviewees by Gender.

percent100 = df.Gender.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
## Frequency of values as percentage ##
## Distribution of surveys by Main Locations.

percent100 = df.City.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
## Frequency of values as percentage ##
## Distribution of surveys by Sub-locations.

percent100 = df.Neighborhood.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
## Frequency of values as percentage ##
## Distribution of surveys within Sub-locations.

ByCity = df.groupby(['City'])

percent100 = ByCity.Neighborhood.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
## Frequency of values as percentage ##
## Distribution of sex of interviewees within Sub-Locations.

ByNbh = df.groupby(['Neighborhood'])

percent100 = ByNbh.Gender.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
## Frequency of values as percentage ##
## Distribution of sex of interviewees by Main Locations.

ByCity = df.groupby(['City'])

percent100 = ByCity.Gender.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
## Frequency of values as percentage ##
## Distribution of Sex of Interviewees within Sub-Locations.

ByCityNbh = df.groupby(['City', 'Neighborhood'])

percent100 = ByCityNbh.Gender.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
ByCityNbh = df.groupby(['City','Neighborhood',])
s = ByCityNbh
counts = s.Gender.value_counts()
PercentT = s.Gender.value_counts()
percent100 = s.Gender.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
pd.DataFrame({'counts': counts, 'percentage': percent100})
```
